Höhe h des Baumes $$ h = log_2 (b) $$  
*i.e., if the number of elements doubles, the tree becomes only one level deeper*  
  
Anzahl Blätter b des Baumes $$ b = {(n+1) \over 2 } = 2^{h} $$
Anzahl Knoten n $$ n = 2^{h+1}-1 $$
Anzahl innere Knoten i (nicht-Blätter): $$ i = 2^{h} -1 $$
Anzahl Knoten auf Level *l*: $$ 2^l $$

```python
def heap_sort(lst):
    n = len(lst)
    heapify(lst)
    print("Heap sorted")
    for end in range(n-1, 0, -1):
        lst[0], lst[end] = lst[end], lst[0]
        sift_down(lst, 0, end - 1)  # end -1, da auf end das geswappte element
    return lst

def heapify(lst):
    """stellt mit gegebener Liste einen gueltigen Heap her,
    die kleinsten Elemente befinden sich unten, das Maximum oben:
    geordneter Baum
    """
    n = len(lst)
    for i in range((n - 2)//2, -1, -1):  # -1 da stop exklusive
        print(f"for_1 {i}")
        sift_down(lst, i, n - 1)
        #display(build(lst))
    
def sift_down(lst, start, end):
    """kleinere Elemente nach unten sickern lassen,
    dabei sind root, child1 und child2 Listenindizes
    und end das Ende der unsortierten Liste
    """
    root = start
    while (root * 2 + 1 <= end):
        print(f"while1: root {root}")
        child1 = root * 2 + 1
        condition = root * 2 + 2
        child2 = condition if condition <= end else None
        
        if child2 is not None:
            condition = lst[child1] > lst[child2]
            max_child = child1 if condition else child2
        else:
            max_child = child1
        
        if lst[max_child] < lst[root]:
            return  ## break ?
        lst[root], lst[max_child] = lst[max_child], lst[root]
        root = max_child
```

```python
# other version of heap sort

def heapify_1(heap, length, parentPos):
    while (True):
        print("heapify ", heap, parentPos)
        leftChildPos = parentPos * 2 + 1
        rightChildPos = parentPos * 2 + 2

        # Find the largest element
        largestPos = parentPos
        if (leftChildPos < length and heap[leftChildPos] > heap[largestPos]):
            largestPos = leftChildPos
        
        if (rightChildPos < length and heap[rightChildPos] > heap[largestPos]):
            largestPos = rightChildPos;
        
        # largestPos is now either parentPos, leftChildPos or rightChildPos.
        # If it's the parent, we're done
        if (largestPos == parentPos):
            break

        # If it's not the parent, then switch!
        heap[parentPos], heap[largestPos] = heap[largestPos], heap[parentPos] 
        # ... and fix again starting at the child we moved the parent to
        parentPos = largestPos

def buildHeap_1(elements):
    # "Find" the last parent node
    lastParentNode = int(len(elements) / 2) - 1

    # Now heapify it from here on backwards
    # (int i = lastParentNode; i >= 0; i--) {
    for i in range(lastParentNode, -1, -1):  
        heapify_1(elements, len(elements), i)

def heapSort_1(elements):
    buildHeap_1(elements)
    print("Heap initialized")
    # for (int swapToPos = elements.length - 1; swapToPos > 0; swapToPos--) {
    for swapToPos in range(len(elements)-1, 0, -1):
        # Move root to end
        elements[0], elements[swapToPos] = elements[swapToPos], elements[0]

        # Fix remaining heap
        heapify_1(elements, swapToPos, 0)
        print("Heap fixed")
    
arr2 = [0,1,2,3,4,5,6]
heapSort_1(arr2)
arr2
```
