```python
def f(x):
    return np.exp(-x)*np.sin(2*np.pi*x)

def g(x):
    return np.exp(-2*x)*np.sin(4*np.pi*x)

n=101
x = np.linspace(0,1,n)

y1 = f(x)
y2 = g(x)

plt.plot(x, y1,x, y2)
#plt.plot(x, y1, 'g:', label = 'exp(-x)*sin(2$\pi$ x)')
#plt.plot(x, y2, 'b--', label = 'exp(-2x)*sin(4$\pi$ x)')

plt.xlabel('x')
plt.ylabel('y')
#plt.legend()

plt.show()

```
